```python
!pip install yfinance
!pip install pyautogui
!pip install pyperclip

import pyautogui
import pyperclip
import yfinance as yf

ticker = input('Digite o código da ação: ')
dados = yf.Ticker(ticker) #ticker é o código de uma ação no yf

tabela = dados.history('6mo') #para coletar os dados da ação num determinado periodo. 
# ano: y, mes: mo, dia: d

#selecionando apenas a coluna de fechamento
fechamento = tabela.Close

fechamento.plot() # gera um gráfico de linha

# Gerando estatisticas
maxima = fechamento.max()
minima = fechamento.min()
atual = fechamento[-1]

print(maxima)
print(minima)
print(atual)

#criando textos do e-mail
destinatario = 'thaelenml15@gmail.com'
assunto = 'Teste de automação'
mensagem = f"""
Bom dia,
Segue abaixo as análises da ação {ticker} nos ultimos 6 meses:

Cotação máxima: R${round(maxima, 2)}
Cotação minima: R${round(minima, 2)}
Contação atual: R${round(atual, 2)}

Atenciosamente,
Eu mesmo.
"""

#automatizando o envio
# configurar pausa entre as açoes do pyautogui
pyautogui.PAUSE = 4

#abrir uma nova aba
pyautogui.hotkey('ctrl', 't')

#copiar o endereço do gmail para o clipboard
pyperclip.copy('www.gmail.com')

#colar o endereço e dar um enter
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')

#clicar no botao escrever
pyautogui.click(x=131, y=332)

#preenchendo o destinatário
pyperclip.copy(destinatario)
pyautogui.hotkey('ctrl','v')
pyautogui.press('tab')

#preenchendo o assunto
pyperclip.copy(assunto)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab')

#preenchendo o corpo do e-mail
pyperclip.copy(mensagem)
pyautogui.hotkey('ctrl','v')

#clicar no botao enviar
pyautogui.click(x=1057, y=965)

#fechar a guia do gmail
pyautogui.hotkey('ctrl','f4')

#mensagem de sucesso
print('E-mail enviado com sucesso!')
```
